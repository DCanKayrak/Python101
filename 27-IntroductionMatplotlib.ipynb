```python
myFigure = plt.figure()
myPlot1 = myFigure.add_axes([0.1,0.1,0.7,0.7])
myPlot2 = myFigure.add_axes([0.2,0.4,0.3,0.3])

myPlot1.plot(npDizisi,npDizisi2,"b")
myPlot1.set_xlabel("X Ekseni")
myPlot1.set_ylabel("Y Ekseni")
myPlot1.set_title("Ana Grafik")

myPlot2.plot(npDizisi2,npDizisi,"y")
myPlot2.set_xlabel("X Ekseni")
myPlot2.set_ylabel("Y Ekseni")
myPlot2.set_title("Yavru Grafik")
```

```python
(benimFiguree,benimEksen) = plt.subplots()
benimEksen.plot(npDizisi,npDizisi2,"b--")
```

```python
(benimFigureee,benimEksenler) = plt.subplots(nrows=1,ncols=2)

for item in benimEksenler:
    item.plot(npDizisi,npDizisi2,"b")
    item.set_xlabel("X Ekseni")
    item.set_ylabel("Y Ekseni"),
plt.tight_layout() # aralarında mesafe açar.
# benimFigureee
```

```python
yeniFigure = plt.figure()

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(npDizisi,npDizisi ** 2,label = "npDizisi ** 2")
yeniEksen.plot(npDizisi,npDizisi ** 3,label = "npDizisi ** 3")
yeniEksen.legend()
```
