```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpy1 = np.linspace(0,10,20)
numpy1
```

```python
numpy2 = numpy1 ** 2
numpy2
```

```python
(benimFigure,benimEksen) = plt.subplots()
benimEksen.plot(numpy1,numpy2,color="#1526",alpha = 0.7)
benimEksen.plot(numpy2,numpy1,color="#7532",alpha = 0.7);
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpy1,numpy1 + 2,"b",linewidth = 5)
yeniEksen.plot(numpy1,numpy1 + 3,"y",linewidth = 3)

yeniEksen.plot(numpy1,numpy1 + 4,"r",linestyle = "-.")
yeniEksen.plot(numpy1,numpy1 + 1,"c",linestyle = "--")

yeniEksen.plot(numpy1,numpy1 + 5,"black",linestyle = "--",marker = "o",markersize = 8,markerfacecolor="red")
```
